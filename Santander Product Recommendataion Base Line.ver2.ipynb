범주형(categorical), 수치형(numerical) 데이터를 전처리합니다.  
범주형 데이터는 .factorize()를 통해 Label Encoding을 수행합니다.  
데이터 타입이 object로 표현되는 수치형 데이터에서는 .unique()를 통해 특이 값들을 대체하거나 제거하고, 정수형 데이터로 변환해 줍니다.

train set과 test set의 columns를 한번에 전처리 하기 위해 통합 시도

```python
df = pd.concat([train, test], axis=0)
```

```python
df
```

test datasets에는 상품 정보(24: 부분의 columns)가 없어서 모두 NaN 값으로 들어감

```python
train.columns[22:]
```

테스트 데이터에 없는 상품 정보 columns는 0으로 대체

```python
for col in train.columns[22:]:
    test[col] = 0
df = pd.concat([train, test], axis=0)
```

```python
df
```

```python
# 학습에 사용할 변수를 담을 list 생성
features = []

# 범주형 변수를 .factorize() 함수를 통해 label encoding
categorical_cols = ['ind_empleado', 'pais_residencia', 'sexo', 'ind_nuevo', 'indrel',
                    'indrel_1mes', 'tiprel_1mes', 'indresi', 'indext', 'conyuemp', 'canal_entrada',
                    'indfall', 'cod_prov', 'ind_actividad_cliente', 'segmento']
for col in categorical_cols:
    df[col], _ = df[col].factorize()
features += categorical_cols
```

```python
df.info()
```

날짜 데이터가 아닌 'age' 'antiguedad' 'renta'는 결측값을 -99로 대체

```python
df.age.unique()
```

```python
# 수치형 변수의 특이값과 결측값을 -99로 대체하고, 정수형으로 변환한다
df['age'].replace(' NA', -99, inplace=True)
df['age'] = df['age'].astype('int')
```

```python
df['antiguedad'].unique()
```

```python
df['antiguedad'].replace('     NA', -99, inplace=True)
df['antiguedad'] = df['antiguedad'].astype('int')
```

'renta'의 경우 unique 값이 너무 많아서 리스트로 받아 확인해 봤더니 nan값이 섞여 있어서  
nan만 수정해주면 될 줄 알았으나, nan 말고도 결측치가 있었음.  
또한 바로 int형으로 전환이 되지 않고 float형으로 변환 후 다시 int로 변환해 주어야 했음

```python
df['renta'].unique()
```

```python
df['renta'].fillna(-99, inplace=True)
df['renta'] = df['renta'].astype('int')
```

```python
df['renta'].replace('         NA', -99, inplace=True)
df['renta'].fillna(-99, inplace=True)
df['renta'] = df['renta'].astype('float').astype('int')
```
